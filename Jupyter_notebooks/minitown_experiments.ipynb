# Experiment data using Minitown and MiniCPS
These experiments were run using minitown and miniCPS, two scenarios are simulated: demand driven and pressure driven. For each scenario, results without attack and results with attacks were simulated. Five attacks were simulated: attack on PLC1, attack on PLC2, man in the middle attack on SCADA, man in the middle attack on PLC2, denial of service attack on PLC2

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Baseline simulation using WNTR and no minicps (works)
no_minicps_data = pd.read_csv('no_minicps.csv')
demand_driven_no_attack = pd.read_csv('experiment_data/demand_driven/no_attack/physical_results.csv')
demand_driven_no_attack_sleep_0_5_second = pd.read_csv('experiment_data/demand_driven/no_attack/sleep_0_5_second/physical_results_sleep_0_5.csv')
demand_driven_mitm_plc2 = pd.read_csv('experiment_data/demand_driven/attack_mitm_plc2/physical_results.csv')
demand_driven_no_attack_sleep_1_second = pd.read_csv('experiment_data/demand_driven/no_attack/sleep_1_second/physical_results.csv')
demand_driven_mitm_plc2_sleep_1_second = pd.read_csv('experiment_data/demand_driven/attack_mitm_plc2/sleep_1_second/physical_results.csv')
demand_driven_mitm_plc2_sleep_0_5_second = pd.read_csv('experiment_data/demand_driven/attack_mitm_plc2/sleep_0_5_second/physical_results_mitm_sleep_0_5.csv')
```

```python
dd = plt.subplot(2,2,1)
demand_driven_no_attack["TANK_LEVEL"].plot(label='DD - No attack', ax=dd)
demand_driven_mitm_plc2["TANK_LEVEL"].plot(label='DD - Mitm PLC2', ax=dd)
plt.title("MiniCPS simulation, running SCADA and 0.5 sleep time")
plt.legend()
plt.subplots_adjust(top=2.00, right=3.00)
plt.grid(True)

dd = plt.subplot(2,2,2)
demand_driven_no_attack_sleep_1_second["TANK_LEVEL"].plot(label='DD - No attack', ax=dd)
demand_driven_mitm_plc2_sleep_1_second["TANK_LEVEL"].plot(label='DD - Mitm PLC2', ax=dd)

plt.title("MiniCPS simulation, running scada and 1 sleep time")
plt.legend()
plt.subplots_adjust(top=2.00, right=3.00)
plt.grid(True)

dd = plt.subplot(2,2,3)
demand_driven_mitm_plc2_sleep_0_5_second["TANK_LEVEL"].plot(label='DD - Mitm PLC2 0.5', ax=dd)
demand_driven_no_attack_sleep_0_5_second["TANK_LEVEL"].plot(label='DD - No attack 0.5', ax=dd)
plt.title("Re simulation with 0.5 sleep time")
plt.legend()
plt.subplots_adjust(top=2.00, right=3.00)
plt.grid(True)


dd = plt.subplot(2,2,4)
no_minicps_data["PRESSURE_TANK"].plot(label='WNTR - No attack', ax=dd)
plt.title("WNTR Simulation")
plt.legend()
plt.subplots_adjust(top=2.00, right=3.00)
plt.grid(True)
```
