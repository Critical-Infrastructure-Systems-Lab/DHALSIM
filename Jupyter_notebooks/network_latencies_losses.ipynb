```python
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns
import math
```

```python
mean_delay = 50.0
sigma_delay = math.sqrt( 2*math.pow(mean_delay,2) / math.pi )
sns.distplot(random.rayleigh(scale=sigma_delay, size=10000), hist=False)
plt.show()
```

```python
delays = random.rayleigh(scale=sigma_delay, size=100)
plt.hist(delays)
```

```python
mean_loss = 2
sigma_loss = math.sqrt( 2*math.pow(mean_loss,2) / math.pi )
sns.distplot(random.rayleigh(scale=sigma_loss, size=100), hist=False)
plt.show()
```

```python
losses = random.rayleigh(scale=sigma_loss, size=100)
plt.hist(losses)
```
