# SUTD WADI Plots
This notebook plots the experiments done in the SUTD WADI simulation
Three plots are presented:
- The basic WADI behavior without attacks
- The behaviour with an attack on the PLC1. During this attack, the PLC1 delays the activation of the actuators
- The behaviour of a man in the middle attack on the PLC1. During this attack, PLC1 receives spoofed information that tells him that tank 2 level is higher than it actually is

```python
import matplotlib.pyplot as plt
import wntr
import wntr.network.controls as controls
import sqlite3
import csv
import time
import pandas as pd
from platform import python_version
```
