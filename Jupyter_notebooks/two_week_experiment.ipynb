```python
import pandas as pd
import matplotlib.pyplot as plt
import statistics
```

```python
no_minicps_data = pd.read_csv('no_minicps.csv')
demand_driven_no_attack = pd.read_csv('experiment_data/demand_driven/no_attack/physical_process.csv')
demand_driven_mitm_plc2 = pd.read_csv('experiment_data/demand_driven/attack_mitm_plc2/physical_results.csv')

pressure_driven_no_attack = pd.read_csv('experiment_data/pressure_driven/no_attack/physical_process.csv')
no_minicps_data_pdd = pd.read_csv('no_minicps_pdd.csv')

demand_driven_diff_minicps_wntr_no_attack = []
for i in range((len(no_minicps_data['TANK_LEVEL']))):
    demand_driven_diff_minicps_wntr_no_attack.append(abs ( no_minicps_data['TANK_LEVEL'][i] - demand_driven_no_attack['TANK_LEVEL'][i]  ) )
    
    
pressure_driven_diff_minicps_wntr_no_attack = []
for i in range((len(no_minicps_data['TANK_LEVEL']))):
    pressure_driven_diff_minicps_wntr_no_attack.append(abs ( no_minicps_data['TANK_LEVEL'][i] - pressure_driven_no_attack['TANK_LEVEL'][i]  ) )
```

```python
dd = plt.subplot(2,1,1)
no_minicps_data["TANK_LEVEL"].plot(label='WNTR - No attack', ax=dd)
demand_driven_no_attack["TANK_LEVEL"].plot(label='DD - No attack', ax=dd)

plt.title("MiniCPS simulation - Demand Driven. Minitown topology")
plt.legend()
plt.xlabel('iterations')
plt.ylabel('tank level (m)')
plt.subplots_adjust(top=2.00, right=3.00)
plt.grid(True)

dd = plt.subplot(2,1,2)
plt.plot(demand_driven_diff_minicps_wntr_no_attack, label="DD - No attack")
plt.title("Difference over time between WNTR and MiniCPS - No attack")
plt.legend()
plt.xlabel('iterations')
plt.ylabel('tank level (m)')
plt.xlim(left=0, right=672)
plt.subplots_adjust(top=2.00, right=3.00)
plt.grid(True)
```

```python
dd = plt.subplot(2,1,1)
no_minicps_data_pdd["TANK_LEVEL"].plot(label='WNTR PDD - No attack', ax=dd)
pressure_driven_no_attack["TANK_LEVEL"].plot(label='MiniCPS PDD - No attack', ax=dd)
plt.title("MiniCPS simulation. Pressure Driven. Minitown topology")
plt.xlabel('iterations')
plt.ylabel('tank level (m)')
plt.legend()
plt.subplots_adjust(top=2.00, right=3.00)
plt.grid(True)

dd = plt.subplot(2,1,2)
plt.plot(demand_driven_diff_minicps_wntr_no_attack, label="PD - No attack")
plt.title("Difference over time between WNTR and MiniCPS - No attack")
plt.legend()
plt.xlabel('iterations')
plt.ylabel('tank level (m)')
plt.xlim(left=0, right=672)
plt.subplots_adjust(top=2.00, right=3.00)
plt.grid(True)
```

```python
dd = plt.subplot(2,1,1)
plt.plot(demand_driven_diff_minicps_wntr_no_attack, label="DD - No attack")
plt.title("Difference over time between WNTR and MiniCPS - No attack")
plt.legend()
plt.xlabel('iterations')
plt.ylabel('tank level (m)')
plt.xlim(left=0, right=672)
plt.subplots_adjust(top=2.00, right=3.00)
plt.grid(True)

dd = plt.subplot(2,1,2)
plt.plot(demand_driven_diff_minicps_wntr_no_attack, label="PD - No attack")
plt.title("Difference over time between WNTR and MiniCPS - No attack")
plt.legend()
plt.xlabel('iterations')
plt.ylabel('tank level (m)')
plt.xlim(left=0, right=672)
plt.subplots_adjust(top=2.00, right=3.00)
plt.grid(True)
```

```python
two_weeks_no_minicps_data = pd.read_csv('two_weeks_no_minicps.csv')
two_weeks_demand_driven_no_attack = pd.read_csv('experiment_data/two_weeks_results/demand_driven/no_attack/physical_process.csv')
two_weeks_pressure_driven_no_attack = pd.read_csv('experiment_data/two_weeks_results/pressure_driven/no_attack/physical_process.csv')
```

```python
dd = plt.subplot(2,1,1)
two_weeks_no_minicps_data["PRESSURE_TANK"].plot(label='WNTR - No attack', ax=dd)
two_weeks_demand_driven_no_attack["TANK_LEVEL"].plot(label='DD - No attack', ax=dd)
plt.title("MiniCPS simulation - Demand Driven. Minitown topology. Two weeks")
plt.legend()
plt.xlabel('iterations')
plt.ylabel('tank level (m)')
plt.xlim(left=0, right=1342)
plt.subplots_adjust(top=2.00, right=3.00)
plt.grid(True)

dd = plt.subplot(2,1,2)
two_weeks_pressure_driven_no_attack["TANK_LEVEL"].plot(label='PD - No attack', ax=dd)
plt.title("MiniCPS simulation - Pressure Driven. Minitown topology. Two weeks")
plt.legend()
plt.xlabel('iterations')
plt.ylabel('tank level (m)')
plt.xlim(left=0, right=1342)
plt.subplots_adjust(top=2.00, right=3.00)
plt.grid(True)
```

```python
two_weeks_demand_driven_diff_minicps_wntr_no_attack = []
for i in range((len(two_weeks_demand_driven_no_attack['TANK_LEVEL']))):
    two_weeks_demand_driven_diff_minicps_wntr_no_attack.append(abs ( two_weeks_no_minicps_data['PRESSURE_TANK'][i] - two_weeks_demand_driven_no_attack['TANK_LEVEL'][i]  ) )
    
two_weeks_pressure_driven_diff_minicps_wntr_no_attack = []
for i in range((len(two_weeks_pressure_driven_no_attack['TANK_LEVEL']))):
    two_weeks_pressure_driven_diff_minicps_wntr_no_attack.append(abs ( two_weeks_no_minicps_data['PRESSURE_TANK'][i] - two_weeks_pressure_driven_no_attack['TANK_LEVEL'][i]  ) )
```

```python
dd = plt.subplot(2,1,1)
plt.plot(two_weeks_demand_driven_diff_minicps_wntr_no_attack, label="DD - No attack")
plt.title("Difference over time between WNTR and MiniCPS - No attack")
plt.legend()
plt.xlabel('iterations')
plt.ylabel('tank level (m)')
plt.subplots_adjust(top=2.00, right=3.00)
plt.xlim(left=0, right=1342)
plt.grid(True)

dd = plt.subplot(2,1,2)
plt.plot(two_weeks_pressure_driven_diff_minicps_wntr_no_attack, label="PD - No attack")
plt.title("Difference over time between WNTR and MiniCPS - No attack")
plt.legend()
plt.xlabel('iterations')
plt.ylabel('tank level (m)')
plt.xlim(left=0, right=1342)
plt.subplots_adjust(top=2.00, right=3.00)
plt.grid(True)
```

```python
demand_one_week_mean = statistics.mean(demand_driven_diff_minicps_wntr_no_attack)
demand_one_week_stdev = statistics.stdev(demand_driven_diff_minicps_wntr_no_attack)

pressure_one_week_mean = statistics.mean(pressure_driven_diff_minicps_wntr_no_attack)
pressure_one_week_stdev = statistics.stdev(pressure_driven_diff_minicps_wntr_no_attack)

demand_two_weeks_mean = statistics.mean(two_weeks_demand_driven_diff_minicps_wntr_no_attack)
demand_two_weeks_stdev = statistics.stdev(two_weeks_demand_driven_diff_minicps_wntr_no_attack)

pressure_two_weeks_mean = statistics.mean(two_weeks_pressure_driven_diff_minicps_wntr_no_attack)
pressure_two_weeks_stdev = statistics.stdev(two_weeks_pressure_driven_diff_minicps_wntr_no_attack)


print("One Week DD: Mean: " + str(demand_one_week_mean)
     + "+/-" + str(demand_one_week_stdev))

print("One Week PD: Mean: " + str(pressure_one_week_mean)
     + "+/-" + str(pressure_one_week_stdev))

print("Two Weeks DD: Mean: " + str(demand_two_weeks_mean)
     + "+/-" + str(demand_two_weeks_stdev))

print("Two Weeks PD: Mean: " + str(pressure_two_weeks_mean)
     + "+/-" + str(pressure_two_weeks_stdev))
```
