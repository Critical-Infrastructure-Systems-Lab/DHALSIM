```python
import matplotlib.pyplot as plt
import wntr
import wntr.network.controls as controls
import sqlite3
import csv
import time
import pandas as pd
```

```python
ctown_epanet = pd.read_csv('epanet/week_0_new/physical_process.csv')
ctown_epanet_no_controls = pd.read_csv('epanet_minicps_no_controls/week_0/physical_process.csv')
```

```python
def do_plot(dd, series, legends, title, labels, xlim=None, ylim=None):
    subplot_hspace = 0.6
    subplot_wspace = 0.1      
    i = 0
    for serie in series:
        serie.plot(label=legends[i], ax=dd)        
        i = i+1
    plt.title(title)
    plt.xlabel(labels[0])
    plt.ylabel(labels[1])
    if xlim and ylim:               
        plt.xlim(left=xlim[0], right=xlim[1])
    plt.legend()
    plt.subplots_adjust(top=2.00, right=3.00, hspace = subplot_hspace, wspace=subplot_wspace)
    plt.grid(True)    
```

```python
dd = plt.subplot(4,2,1)
do_plot(dd,[ctown_epanet["T1_LEVEL"], ctown_epanet_no_controls["T1_LEVEL"]],
        ['EPANET-PDA - MiniCPS', 'EPANET-PDA - No controls'],'CTown Topology - Tank 1',['iterations', 'tank level (m)'])
dd = plt.subplot(4,2,2)
do_plot(dd,[ctown_epanet["PU1_STATUS"], ctown_epanet_no_controls["PU1_STATUS"]],
        ['EPANET-PDA - MiniCPS', 'EPANET-PDA - No controls'],'CTown Topology - PU1 Status',['iterations', 'status'])
dd = plt.subplot(4,2,3)
do_plot(dd,[ctown_epanet["PU2_STATUS"], ctown_epanet_no_controls["PU2_STATUS"]],
        ['EPANET-PDA - MiniCPS', 'EPANET-PDA - No controls'],'CTown Topology - PU2 Status',['iterations', 'status'])
```
